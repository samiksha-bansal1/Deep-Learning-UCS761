```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

### AND Gate

```python
#dataset
x=np.array([[0,0],[0,1],[1,0],[1,1]],dtype=np.float32)
y=np.array([0,0,0,1],dtype=np.float32)
df=pd.DataFrame({'X1': x[:,0], 'X2': x[:,1], 'Y': y})
print(df)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

#perceptron
model_AND = Sequential()
model_AND.add(Dense(1,input_dim=2,activation='sigmoid'))
```

```python
from tensorflow.keras.optimizers import Adam
model_AND.compile(loss='binary_crossentropy',optimizer=Adam(learning_rate=0.1),metrics=['accuracy'])
```

```python
model_AND.fit(x,y,epochs=50)
```

```python
print((model_AND.predict(x)>0.5).astype(int))
```

```python
model_AND.layers
```

```python
weights,biases=model_AND.layers[0].get_weights()
print(weights)
print(biases)
```

### OR Gate


```python
x=np.array([[0,0],[0,1],[1,0],[1,1]],dtype=np.float32)
y=np.array([0,1,1,1],dtype=np.float32)
df=pd.DataFrame({'X1': x[:,0], 'X2': x[:,1], 'Y': y})
print(df)

model_OR=Sequential()
model_OR.add(Dense(1,input_dim=2,activation='sigmoid'))
print(model_OR.summary())
model_OR.compile(loss='binary_crossentropy',optimizer=Adam(learning_rate=0.1),metrics=['accuracy'])
model_OR.fit(x,y,epochs=50)
```

```python
model_OR.layers[0].get_weights()
```

```python
print((model_OR.predict(x)>0.5).astype(int))
```

###NOT Gate

```python
x=np.array([0,1],dtype=np.float32)
y=np.array([1,0],dtype=np.float32)

model_NOT=Sequential()
model_NOT.add(Dense(1,input_dim=1,activation='sigmoid'))
print(model_NOT.summary())
model_NOT.compile(loss='binary_crossentropy',optimizer=Adam(learning_rate=0.1),metrics=['accuracy'])
model_NOT.fit(x,y,epochs=10)

print(model_NOT.layers[0].get_weights())
print((model_NOT.predict(x)>0.5).astype(int))
```

###XOR

```python
#as xor is not linearly seperable so we need MLP
x=np.array([[0,0],[0,1],[1,0],[1,1]],dtype=np.float32)
y=np.array([[0],[1],[1],[0]],dtype=np.float32)

model_XOR=Sequential()
model_XOR.add(Dense(2,input_dim=2,activation='sigmoid'))
model_XOR.add(Dense(1,activation='sigmoid'))
print(model_XOR.summary())

model_XOR.compile(optimizer=Adam(learning_rate=0.1),loss='binary_crossentropy',metrics=['accuracy'])
model_XOR.fit(x,y,epochs=100)

print(model_XOR.layers[0].get_weights())
print((model_XOR.predict(x)>0.5).astype(int))
```
