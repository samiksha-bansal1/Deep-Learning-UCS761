```python
n_features=X.shape[1]
w=np.zeros(n_features)
b=0.0
```

```python
# 1.Why do we need one weight per feature?
# Ans: one for each input feature so that we can adjust the weightage /role of that feature and to control the impact of it on the output variable.

# 2.Why is bias separate?
# Ans: bias is separate because it adds non linearity to the model and its used to shift the decision boundary.

# 3.Would initializing with large values be risky
# Ans : it can lead to large errors in the predictions.
```

```python
def predict(X, w, b):
    """
    X : input features
    w : weights
    b : bias
    returns predicted values
    """
    y_hat = X.dot(w) + b
    return y_hat
```

```python
def mean_squared_error(y, y_hat):
  """
  y     : actual values
  y_hat : predicted values
  """
  loss = ((y_hat - y) **2).mean()
  return loss
```

```python
def compute_gradients(X, y, y_hat):
  """
  Compute gradients of MSE loss
  """
  N =len(y)
  # Gradient w.r.t weights
  dw = (2 / N) * X.T.dot(y_hat - y)
  # Gradient w.r.t bias
  db = (2 / N) * (y_hat - y).sum()
  return dw, db
```

```python
def update_parameters(w, b, dw, db, lr):
  w = w - lr * dw
  b = b - lr * db
  return w, b
```

```python
lr =0.0001
epochs =1000
```

```python
for epoch in range(epochs):
  y_hat = predict(X, w, b)
  loss = mean_squared_error(y, y_hat)
  dw, db = compute_gradients(X, y, y_hat)
  w, b = update_parameters(w, b, dw, db, lr)
  if epoch %100 ==0:
    print(f"Epoch {epoch}, Loss:{loss}")
```

```python
new_candidate = np.array([4.5,68])
predicted_salary = new_candidate.dot(w) + b
print(predicted_salary)
```
