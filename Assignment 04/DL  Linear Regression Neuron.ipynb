```python
#train-test-split

import random

def traintestsplit(X,y,testsize=0.3,randomseed=None):
    data=list(zip(X,y))

    if randomseed is not None:
        random.seed(randomseed)
    
    random.shuffle(data)

    idx=int(len(data)*(1-testsize))
    train=data[:idx]
    test=data[idx:]

    X_train,y_train=zip(*train)
    X_test,y_test=zip(*test)

    return (np.array(X_train),np.array(X_test),np.array(y_train),np.array(y_test) )   

```

```python
X_train,X_test,y_train,y_test=traintestsplit(X,y,testsize=0.2,randomseed=42)
```

```python
class StandardScaler:
  def __init__(self):
    self.mean=None
    self.std=None

  def fit(self,X):
    X=np.array(X)
    self.mean=X.mean(axis=0)
    self.std=X.std(axis=0)

  def transform(self,X):
    X=np.array(X)
    X_scaled=(X-self.mean)/self.std
    return X_scaled

  def fit_transform(self,X):
    X=np.array(X)
    self.fit(X)
    return self.transform(X)

  def inverse_transform(self,X_scaled):
    X_scaled=np.array(X_scaled)
    X=X_scaled*self.std+self.mean
    return X
```

```python
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
def forward (X,w,b):
    y_hat=np.dot(X,w)+b
    return y_hat
```

```python
def mse(y,y_hat):
    loss=np.mean((y-y_hat)**2)
    return loss
```

```python
def grad_w(X,y,y_hat):
    N=len(y)
    dW=(-2/N)*np.dot(X.T,(y-y_hat))
    return dW

def grad_b(y,y_hat):
    N=len(y)
    db=(-2/N)*np.sum(y-y_hat)
    return db
```

```python
d=X_train.shape[1]
print(d)
```

```python
w=np.random.randn(d,1)*0.02
w
```

```python
b=0
```

```python
lr=0.01
epochs=1000

for epoch in range(epochs):
    y_hat=forward(X_train,w,b)

    loss=mse(y_train,y_hat)

    dw=grad_w(X_train,y_train,y_hat)
    db=grad_b(y_train,y_hat)

    w=w-lr*dw
    b=b-lr*db
    
    if epoch%100==0:
        print(f"epoch {epoch},loss= {loss}")
```

```python
y_test_hat=np.dot(X_test, w) + b
test_mse=np.mean((y_test - y_test_hat)**2)
test_mae=np.mean(np.abs(y_test - y_test_hat))

print("Test MSE:",test_mse)
print("Test MAE:",test_mae)
```
